```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('AirQualityUCI.xlsx',parse_dates=[['Date','Time']])
df
```

```python
df.dtypes
```

```python
data=df.drop(columns='Date_Time')
data.index=df['Date_Time']
data
```

```python
cols=data.columns
for j in cols:
    for i in range(len(data)):
        if data[j][i]==-200:
            data[j][i]=data[j][i-1]
```

```python
data
```

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen
johan_test_temp=data.drop(['CO(GT)'],axis=1)
coint_johansen(johan_test_temp,-1,1).eig
```

```python
train=data[:int(0.8*(len(data)))]
valid=data[int(0.8*(len(data))):]
```

```python
from statsmodels.tsa.vector_ar.var_model import VAR
model=VAR(endog=train)
model_fit=model.fit()

prediction=model_fit.forecast(model_fit.y,steps=len(valid))
prediction=pd.DataFrame(prediction,columns=cols)
prediction
```

```python
valid.index=range(0,len(valid))
valid
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
for i in cols:
    print('rmse value for',i,'is: ',sqrt(mean_squared_error(prediction[i],valid[i])))
```
