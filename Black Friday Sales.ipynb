```python
df_train['User_ID'] = df_train['User_ID'] - 1000000
df_test['User_ID'] = df_test['User_ID'] - 1000000
df_train.head()
```

```python
df_test.head()
```

```python
df_train['Product_ID'] = df_train['Product_ID'].str.replace('P00','')
df_test['Product_ID'] = df_test['Product_ID'].str.replace('P00','')

df_train.head()
```

```python
df_train['Age'] = df_train['Age'].map({'0-17': 15,
                               '18-25': 21,
                               '26-35': 30,
                               '36-45': 40,
                               '46-50': 48,
                               '51-55': 53,
                               '55+': 60})
df_test['Age'] = df_test['Age'].map({'0-17': 15,
                               '18-25': 21,
                               '26-35': 30,
                               '36-45': 40,
                               '46-50': 48,
                               '51-55': 53,
                               '55+': 60})
```

```python
df_train['Stay_In_Current_City_Years']=df_train['Stay_In_Current_City_Years'].map({'0':0,'1':1,'2':2,'3':3,'4+':4})
df_train['Stay_In_Current_City_Years']
```

```python
df_test['Stay_In_Current_City_Years']=df_test['Stay_In_Current_City_Years'].map({'0':0,'1':1,'2':2,'3':3,'4+':4})
df_test['Stay_In_Current_City_Years']
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_train['Product_ID']=df_train['Product_ID'].astype('int64')
df_test['Product_ID']=df_test['Product_ID'].astype('int64')
```

```python
df_train.dtypes
```

```python
df_test.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df_train['new']=encoder.fit_transform(df_train['City_Category'])
df_train['new']
```

```python
df_train.drop(columns='new',inplace=True)
df_train
```

```python
train_dummy=pd.get_dummies(df_train,drop_first=True)
train_dummy
```

```python
train_dummy.dtypes
```

```python
test_dummy=pd.get_dummies(df_test,drop_first=True)
test_dummy
```

```python
test_dummy.dtypes
```

# Splitting train data

```python
from sklearn.model_selection import train_test_split
```

```python
X=train_dummy.drop(columns='Purchase')
y=train_dummy['Purchase']
```

# SCALING

```python
from sklearn.preprocessing import StandardScaler
```

Scaling on training data

```python
sc=StandardScaler()
X_new=pd.DataFrame(sc.fit_transform(X))
X_new.columns=X.columns
X_new.index=X.index
```

```python
X_new
```

Scaling on test data

```python
sc=StandardScaler()
test=pd.DataFrame(sc.fit_transform(test_dummy))
test.columns=test_dummy.columns
test.index=test_dummy.index
test
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(X_new,y,test_size=0.2,random_state=21)
```

# Model Training

1)  LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
sqrt(mean_squared_error(pred,y_valid))
```

2) DECISION TREE REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(min_samples_split=10,min_samples_leaf=3,random_state=21)
```

```python
dt.fit(X_train,y_train)
```

```python
pred=dt.predict(X_valid)
```

```python
sqrt(mean_squared_error(pred,y_valid))
```

```python
pred=dt.predict(test)
```

```python
submission=pd.read_csv('sample_submission.csv')
submission['Purchase']=pred
submission.to_csv('DecisionTree.csv',index=False)
```

3) RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=100,random_state=21)
```

```python
rf.fit(X_train,y_train)
```

```python
pred=rf.predict(X_valid)
```

```python
sqrt(mean_squared_error(pred,y_valid))
```

```python
pred=rf.predict(test)
```

```python
submission=pd.read_csv('sample_submission.csv')
submission['Purchase']=pred
submission.to_csv('RandomForest.csv',index=False)
```
